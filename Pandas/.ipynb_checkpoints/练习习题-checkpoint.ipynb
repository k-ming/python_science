### 给定某城市一周的的最高温度Series，完成下面的分析
- 找出温度超过30度的天数
- 计算平均气温
- 将温度从高到低排序
- 找出温度变化最大的两天

```python
import pandas as pd
temperatures = pd.Series([28,31,29,32,30,27,33], index=['一','二','三','四','五','六','七'])
print(temperatures)
```

```python
# 温度超过30天的天数
print(temperatures[temperatures>30].count())
# 换一种方法
print(len(temperatures[temperatures>30]))
```

```python
# 计算平均气温
print(temperatures.mean())
# 换一中方法
print(temperatures.sum()/len(temperatures))
```

```python
# 将温度从高到低排序
sorted = temperatures.sort_values(ascending=False)
print('按值排序:\n',sorted)
print('按索引排序:\n',temperatures.sort_index(ascending=False))
```

```python
# 找出温度变化最大的两天
# 计算相邻温度变化幅度（绝对值）
tem_diff = temperatures.diff().abs()
print(tem_diff)

# 对变化值进行排序
sorted_diif = tem_diff.sort_values(ascending=False)
print(*sorted_diif[0:2].index)
```

## 股票价格分析
- 给定某股票连续10个交易日的收盘价Series:
- 计算每日收益率（当日收盘价/前日收盘价-1）
- 找出收益率最高和最低的日期
- 计算波动率（收益率的标准差）

```python
import pandas as pd
prices = pd.Series([102.3, 103.5, 105.1, 184.8, 106.2, 107.0, 106.5,
108.1, 109.3, 118.2], index=pd.date_range('2023-01-01', periods=10))
print(prices)
```

```python
# 收益率计算 (后日/前一日 - 1)
pct_value = prices.pct_change()
print(pct_value)
print('收益率最高日期:\n', pct_value.idxmax())
print('收益率最低日期:\n', pct_value.idxmin())
print('收益波动率-标准差:\n', pct_value.std())
```

## 销售数据分析
- 某产品过去12个月的销售量Serieg:
- 计算季度平均销量（每3个月为一个季度）
- 找出销量最高的月份
- 计算月环比增长率
- 找出连续增长超过2个月的月份

```python
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190,200, 210,
220],index=pd.date_range('2022-01-01', periods=12,freq='M')) # freq='M'表示以月为单位
print(sales)

# 每季度平均销量,Q表示每季度
avg_amount=sales.resample('Q').mean()
print('每季度平均销量:\n',avg_amount)
    

# 销售量最高的月份
print('销售量最高的月份:\n', sales.idxmax())

# 计算月环比增长率
mon_incre=sales.pct_change()
print('月环比增长:\n',mon_incre)

# 连续增长超过2个月的月份
# 在月环比的基础上，按照3个单位滑动窗口，如果3个单位内的增长都是正数，那么最后一个月就满足
is_incre=mon_incre>0
print('布尔算出是否增涨：\n',is_incre)
#滑动3个单位窗口，并求和，如果和等于3，则满足
rolls=is_incre.rolling(3).sum()
print('滑动窗口求和结果：\n',rolls)
rolls[rolls==3].keys().tolist()
```
